```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
```

### Task 1.

##### a.

```python
df = pd.read_csv('data_assignment2.csv')

# a
x = df['Living_area'].to_numpy().reshape(-1,1)
y = df['Selling_price']
model = LinearRegression().fit(x, y)
```

##### b.

```python
coef = model.coef_
interc = model.intercept_

print(f"Slope: {coef[0]}")
print(f"Intercept: {interc}")
```

#### c.

```python
def calc_price(area):
    return coef * area + interc

areas = [10,100,150,200,1000]

for area in areas:
    print(f"Area: {area}m², Price: {calc_price(area)[0]} kr")
```

#### d.

```python
xfit = np.linspace(0, 10, 300) #1000 evenly spaced points in [0, 10].
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit)
plt.xlabel('Living area [m²]')
plt.ylabel('Price [SEK]')
plt.title('Living area vs Price')
```
